```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/imigrantes_canada.csv")
df.head()
```

```python
df.set_index("País", inplace=True)
df.head()
```

```python
anos = list (map(str, range(1980,2014)))
anos
```

```python
brasil = df.loc["Brasil", anos]
brasil
```

```python
dados_brasil = pd.DataFrame({'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()})
dados_brasil.head()
```
