```python
# Crie um grafo
g = Graph()

# Defina um namespace
ex = Namespace("http://example.com#")

# Adicione um recurso com o namespace
bob = ex["Bob"]

# Adicione uma propriedade com o namespace
g.add((bob, ex["age"], Literal(30)))

# Exiba o grafo
print(g.serialize(format="turtle"))
```
